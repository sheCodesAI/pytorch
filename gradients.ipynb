```python
import torch
```

```python
a = torch.tensor([2., 3.], requires_grad=True)
b = torch.tensor([6., 4.], requires_grad=True)

f = 3 * a**3 - b**2
f
```

Without gradients:

❌ AI cannot learn.
❌ It keeps making the same mistakes.

With gradients:

✅ AI learns from errors.
✅ Accuracy improves after every training step.
In one sentence

requires_grad=True tells PyTorch to remember the calculations so it can later figure out how to improve the AI model during training.

PyTorch is mainly used to train AI models.
PyTorch uses tensors to represent data (images, text, audio, etc.), processes that data in small batches, and updates the model after each batch until it has learned from the entire dataset.

```python
f.backward(gradient = torch.tensor([1,1]))
```

```python
print(a.grad)
```
